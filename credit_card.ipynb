# Training And Spliting DataSet

```python
X= credit.drop( 'Class', axis=1).values
y = credit.Class
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1300)   #test size=20
```

```python
X_train
```

```python
len(X_test)
```

# For RandomForest Classifier

#importing libraries

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# Here we are taking the class_weight to increase the overall performance of the model
For calculating Class_weight,we have the following formula:-
Class_weight=Total number of majority class ie('0')/Total number of minority class ie(1)
In our example,284315/492=int(577.82)=578

```python
w={0:1,1:578}
clf =  RandomForestClassifier(n_estimators=100,class_weight=w)     
```

```python
clf.fit(X_train,y_train)
```

```python
g=clf.predict(X_test)
```

```python
y_test
```

```python
clf.score(X_test,y_test)     #for accuracy
```

```python
print( np.unique( g ) )     # Here np.unique returns 0 and 1 ie the model is predicting both minor and major class 
```

For calculating confusion matrix,precison,recall,roc,avg presicion_score and F1 score


```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix:  ",confusion_matrix(y_test,g))
```

```python
from sklearn.metrics import recall_score
print("Recall:  ",recall_score(y_test,g))
```

```python
from sklearn.metrics import precision_score
print("Precision: ",precision_score(y_test,g))
```

```python
from sklearn.metrics import average_precision_score

print("Average precision-recall score:" ,average_precision_score(y_test,g))
```

```python
from sklearn.metrics import roc_auc_score
print("Area under curve: ",roc_auc_score(y_test,g))
```

```python
from sklearn.metrics import f1_score

print("F1 score:",f1_score(y_test,g))
```

# For Xgboost Classifier

```python
from xgboost import XGBClassifier
```

```python
# Here there are different objective 
# But if we use reg:tweedie it would make precision=1 but will decrease the recall which will decrease the overall F1 score
```

```python

xgb =XGBClassifier( learning_rate=0.1,
 n_estimators=500,
 max_depth=7,
 min_child_weight=1,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'binary:logistic',
 seed=100,
 scale_pos_weight=578)
```

```python
xgb.fit(X_train,y_train)
```

```python
g1=xgb.predict(X_test)
```

```python
xgb.score(X_test,y_test)
```

```python
print( np.unique( g1) )
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix:\n\n ",confusion_matrix(y_test,g1))
```

```python
from sklearn.metrics import recall_score
print("Recall:  ",recall_score(y_test,g1))
```

```python
from sklearn.metrics import precision_score
print("Precision: ",precision_score(y_test,g1))
```

```python
from sklearn.metrics import roc_auc_score
print("Area under curve: ",roc_auc_score(y_test,g1))
```

```python
from sklearn.metrics import average_precision_score

print("Average precision-recall score:" ,average_precision_score(y_test,g1))
```

```python
from sklearn.metrics import f1_score

print("F1 score:",f1_score(y_test,g1))
```

# So in both the models we are getting the same percent of accuracy 
#But we cant really rely on the accuracy as the data is unbalanced
#So here we can consider the F1 score which is the harmonic mean of precision and recall
#i.e your F1 score will be high(â†‘)until and unless both the precision and recall are high(â†‘).


So here XGBoost Classifier will better than Random Forest Classifer.
